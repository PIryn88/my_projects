## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.

missing_values_count = Airbnb_data_dropped.isnull().sum().sort_values(ascending=False)
missing_values_share = missing_values_count / len(Airbnb_data_dropped)

missing_values = pd.concat([missing_values_count, missing_values_share, Airbnb_data_dropped.dtypes], axis = 1)
missing_values.columns = ['count', 'share', 'dtypes']

missing_values
```

```python
# Check host_identity_verified values

pd.unique(Airbnb_data_dropped['host_identity_verified'])
```

In this case it is possible to replace nan-type with an 'unconfirmed' value

```python
# Replace nan-values in column "host_identity_verified" with relevant meanings('unconfirmed') and check it

Airbnb_data_dropped['host_identity_verified'].fillna('unconfirmed', inplace=True)
Airbnb_data_dropped.value_counts('host_identity_verified')
```

```python
# Check the minimum nights

Airbnb_data_dropped['minimum nights'].min()
```

```python
# It's strange to have a number of min values less than 0, so replace these values and nan-values  with 1. 
# Then check the min values again

Airbnb_data_dropped.loc[ Airbnb_data_dropped['minimum nights'] < 0, 'minimum nights'] = 1
Airbnb_data_dropped['minimum nights'].fillna(1, inplace=True)

Airbnb_data_dropped['minimum nights'].fillna(1, inplace=True)
print(Airbnb_data_dropped['minimum nights'].isnull().sum())
```

```python
# Column "instant_bookable" has True/False meaning, so, replace all nan-values with False

Airbnb_data_dropped['instant_bookable'].fillna('FALSE', inplace=True)
print(Airbnb_data_dropped['instant_bookable'].isnull().sum())
```

```python
# For other columns use empty (for object type) or 0, or mean values, or drop them. Then check it

Airbnb_data_dropped["cancellation_policy"].fillna(Airbnb_data_dropped["cancellation_policy"].mode().values[0], inplace=True)
Airbnb_data_dropped["availability 365"].fillna(round(Airbnb_data_dropped["availability 365"].mean(), 1), inplace=True)
Airbnb_data_dropped["calculated host listings count"].fillna(0, inplace=True)
Airbnb_data_dropped["review rate number"].dropna(inplace=True)
Airbnb_data_dropped["neighbourhood"].dropna(inplace=True)
Airbnb_data_dropped["neighbourhood group"].dropna(inplace=True)
Airbnb_data_dropped["cancellation_policy"].dropna(inplace=True)
Airbnb_data_dropped["price"].dropna(inplace=True)
Airbnb_data_dropped["service fee"].dropna(inplace=True)


for column in Airbnb_data_dropped.columns:
    if Airbnb_data_dropped.dtypes is object:
        Airbnb_data_dropped[column].fillna('', inplace=True)
    else:
        Airbnb_data_dropped[column].fillna(0, inplace=True)

Airbnb_data_dropped.isnull().sum()
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.

duplicates_count = Airbnb_data_dropped.duplicated().sum()

if duplicates_count > 0:
    print(f"There are \033[1m{duplicates_count}\033[0m duplicates before removing\n")
```

```python
# Drop the duplicates

Airbnb_data_dropped = Airbnb_data_dropped.drop_duplicates()
```

```python
## Display the total number of records in the dataframe after removing the duplicates.

duplicates_count = Airbnb_data_dropped.duplicated().sum()
print(f"There are \033[1m{duplicates_count}\033[0m duplicates after removing")
```

This data set consists mainly of 'Entire home/apt' and 'Private room' type. The biggest count has the 'Entire home/apt'. There are almost no shared and hotel room types. And this is expected, because Airbnb service more about private home

```python
## Which room type adheres to more strict cancellation policy

df_room_strict_policy = Airbnb_data_dropped[Airbnb_data_dropped['cancellation_policy']=='strict']
df_room_type_counts = df_room_strict_policy.room_type.value_counts()
max_count_room_types = df_room_strict_policy.room_type.value_counts().max()

name_max_room_types = df_room_strict_policy.room_type.value_counts().idxmax()

print(f'{df_room_type_counts} \n\n')
      
print(f'The most often in this data frame with strict cancellation police are \033[1m{name_max_room_types}\033[0m - \033[1m{max_count_room_types}\033[0m')
```

But we don't know the total number of the rooms per type and percent of the rooms with strict cancellation policy in every type.

```python
# Analyze the share of the rooms types

policy_type = Airbnb_data_dropped[Airbnb_data_dropped['cancellation_policy']=='strict']
room_type_counts = Airbnb_data_dropped.room_type.value_counts()
policy_type_counts = policy_type.room_type.value_counts()
df_policy = pd.concat([room_type_counts, policy_type_counts, round(policy_type_counts/room_type_counts*100,2),\
                      round(room_type_counts/len(Airbnb_data_dropped)*100,2)], axis = 1)
df_policy.columns = ['rooms', 'policy', 'share', 'share_rooms_type']

df_policy
```

The first three types of rooms in the table have very similar share values - it's about 33%. Shared room has a little advantage - its share is 33.41%. But the amount of this type of room in the total rooms count is only 2%. Entire home and Private room constitute the main part of the total count. And among these groups the 'private' type has the highest share - 33.27%. To end up, the Private room adheres to a more strict cancellation policy.

```python
# Compare the minimum and maximum price between the groups

price_mean = Airbnb_data_dropped.groupby("neighbourhood_group").price.mean().round(2)
price_min = Airbnb_data_dropped.groupby("neighbourhood_group").price.min().round(2)
price_max = Airbnb_data_dropped.groupby("neighbourhood_group").price.max().round(2)
counts = Airbnb_data_dropped.neighbourhood_group.value_counts()
counts_min = Airbnb_data_dropped.query('price == 50').groupby(["neighbourhood_group"])['price'].count()
counts_max = Airbnb_data_dropped.query('price == 1200').groupby(["neighbourhood_group"])['price'].count()
df_price = pd.concat([price_mean, price_min, price_max, counts, round(counts_min/counts*100,2), round(counts_max/counts*100,2)], axis=1)
df_price.columns=['mean', 'min', 'max', 'counts',  'share_min_price', 'share_max_price']
print(df_price)
```

Interestingly, the Bronx has the largest share of the most expensive offers. The average rental price in Manhattan is not the highest. The share of the cheapest and most expensive offers is about 0.06%. But it is in this group of neighborhoods that there are the most offers, including the most expensive ones.

## Task 5a: Data Visualization (Any Tool)

- Create a horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset.
    - Create another chart with the 10 cheapest neighborhoods in the dataset.
- Create a box and whisker chart that showcases the price distribution of all listings split by room type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshots of your work.

```python
# Create a horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset

df_neighbourhood_prices_high = Airbnb_data_dropped.groupby('neighbourhood')['price'].mean().nlargest(10).reset_index()
sns.barplot(x="price", y="neighbourhood", data=df_neighbourhood_prices_high)
plt.xlabel('Price, $')
plt.ylabel('Neighbourhood name')
plt.title("Top 10 the most expensive neighborhoods")

plt.show()
```

```python
# Create another chart with the 10 cheapest neighborhoods in the dataset.

df_neighbourhood_prices_small = Airbnb_data_dropped.groupby('neighbourhood')['price'].mean().nsmallest(10).reset_index()
sns.barplot(x="price", y="neighbourhood", data=df_neighbourhood_prices_small)
plt.xlabel('Price, $')
plt.ylabel('Neighbourhood name')
plt.title("Top 10 the cheapest neighborhoods")

plt.show()
```

```python
# Create a box and whisker chart that showcases the price distribution of all listings split by room type.

sns.boxplot(x = Airbnb_data_dropped.room_type, y = Airbnb_data_dropped.price)
plt.show()
```

According to this box plot, Shared room has the most variability and average price. Hotel rooms seem the most expensive type. But the main rental offers in this data are private rooms and entire homes. In this case the values are so similar and the difference between their average values is very small.
